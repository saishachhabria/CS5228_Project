```python
import pandas as pd
from sklearn import tree
from sklearn import metrics
from sklearn.preprocessing import OneHotEncoder
```

```python
df_train = pd.read_csv('data/train-cleaned.csv')
```

```python
df_train.head()
```

```python
df_train.drop(['month','street_name','subzone'],axis=1,inplace=True)
```

```python
pd.get_dummies(df_train['town'], prefix='town')
pd.get_dummies(df_train['flat_model'], prefix='flat_model')
pd.get_dummies(df_train['flat_type'], prefix='flat_type')
pd.get_dummies(df_train['eco_category'], prefix='eco_category')
pd.get_dummies(df_train['block'], prefix='block')
pd.get_dummies(df_train['region'], prefix='region')
pd.get_dummies(df_train['planning_area'], prefix='planning_area')
```

```python
ohe = OneHotEncoder()
ohe.fit_transform(df_train['town'].values.reshape(-1, 1)).toarray()
ohe.fit_transform(df_train['flat_model'].values.reshape(-1, 1)).toarray()
ohe.fit_transform(df_train['eco_category'].values.reshape(-1, 1)).toarray()
ohe.fit_transform(df_train['region'].values.reshape(-1, 1)).toarray()
ohe.fit_transform(df_train['planning_area'].values.reshape(-1, 1)).toarray()
ohe.fit_transform(df_train['flat_type'].values.reshape(-1, 1)).toarray()
ohe.fit_transform(df_train['block'].values.reshape(-1, 1)).toarray()
```

```python
df_train = df_train.drop(['town','flat_model','eco_category','region','planning_area','flat_type','block'], axis=1)
```

```python
X= df_train.drop('resale_price',axis=1).values
y=df_train['resale_price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model=Sequential()

model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(X_train,y_train,
          validation_data=(X_test,y_test),
          batch_size=128,epochs=400)
```

```python
losses=pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
import numpy as np
```

```python
pred=model.predict(X_test)
```

```python
mean_absolute_error(y_test,pred)
```

```python
np.sqrt(mean_squared_error(y_test,pred))
```
